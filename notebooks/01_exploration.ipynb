```python
import json, random

path = "../data/US/train.jsonl"
subset = []
with open(path, 'r', encoding='utf-8') as f:
    for line in f:
        q = json.loads(line)
        if "step2" in q["meta_info"]:
            subset.append(q)

mini = random.sample(subset, 100)
json.dump(mini, open("../data/mini_medqa_50.json", "w"), indent=2)
```
